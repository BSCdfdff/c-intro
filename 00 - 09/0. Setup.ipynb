## Setting up the .Net Core environment

____


We create, conda environment and activate it
:

```
conda create -n netcore

conda activate netcore

```

We install miniconda here using conda, and extensions:
```
//Need to install nodejs
conda install -c conda-forge nodejs

//Install conda and extension
conda install -c conda-forge jupyterlab=1 jupyter_conda 
jupyter labextension install jupyterlab_toastify jupyterlab_conda
```

We following the instructions from the folllowing link to install .NETCore SDK (ASPNET Runtime,.NetCore Runtime) for Debian Buster.

https://docs.microsoft.com/en-gb/dotnet/core/install/linux-package-manager-debian10

```
Add Microsoft repository key and feed
-------------------------------------

wget -O - https://packages.microsoft.com/keys/microsoft.asc | gpg --dearmor > microsoft.asc.gpg
sudo mv microsoft.asc.gpg /etc/apt/trusted.gpg.d/
wget https://packages.microsoft.com/config/debian/10/prod.list
sudo mv prod.list /etc/apt/sources.list.d/microsoft-prod.list
sudo chown root:root /etc/apt/trusted.gpg.d/microsoft.asc.gpg
sudo chown root:root /etc/apt/sources.list.d/microsoft-prod.list

Install the .NET Core SDK
-------------------------

sudo apt-get update
May 22, 2020, 10:14:16
sudo apt-get update
sudo apt-get install dotnet-sdk-3.1


Install the ASP.NET Core runtime
--------------------------------

sudo apt-get update
sudo apt-get install apt-transport-https
sudo apt-get update
sudo apt-get install aspnetcore-runtime-3.1


Install the .NET Core runtime
-----------------------------

sudo apt-get update
sudo apt-get install apt-transport-https
sudo apt-get update
sudo apt-get install dotnet-runtime-3.1

```
