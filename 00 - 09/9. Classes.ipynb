Here we going to look at Encapsulation...

When it comes to fields, we want to make their access to the outside protected. Here we use getters and setters. We wil define fields as private.


```
public class Person{

    }

```

## Properties

_____

Its a class member that is encapsulated by ```get``` and ```set``` method.

Completed above, but we going to have a getter adn setter for each field. Isn't there an easier way?

We still have the field, but in addition for field, we define a property with getter and setter methods


```
public class Person{
    private string _name;

    public string Name{
        get { return _name;}
        set { _name = value;}
    }
}
```
Naming: Camel (first letter small) Case for fields, but prefixed with underline.
Naming: Pascal (Caps for letters) Case for Property


But we can even further shorten the above code (auto implemented propertis)

```
   public string Name{get;set;}

```

There is no field, just get and set, the compiler will create an internal field for us.). And Property is defined with public access modifier

So how do we call it? See below.


`````
public class Person{
    public string Name{get;set;}
}
`````

`````
var person = new Person();
person.Name = "Kotlin";
Console.WriteLine("My name is {0}",person.Name);
`````

## Indexers
____

A way to access elements in class that represents a list of values.

If we thinj of an array, and to access an element of array, we use an index....

Why do we need an index for a class, given that an index give you access to elements of group of elements?

So instead of using .(dot) notation to access a member or class, we use ```[index]``` square brackets with member name inside., looks more cleaner

```
var person = new Person()
person["Name"] = "Jackespeare";
```

Bur how do we declare an indexer? An indexer is just a property. We declare an indexer exactly the way we declare a property.

but instead of variable name, we use ```this``` keyword. Followed by brackets, then we specify the data type and variable name in our case the key. But we need to use a Dictionary to implement the key/value concept, to the varaible that will use indexer (in out case name)

```
public class Person{
  private Dictionary<string, string> _name;
  
  public Person(){
      _name = new Dictionary<string, string>();
  
  }
  
  public string this[string key]{
    get { return _name[key];}
    set {  _name[key] = value;}
  }
}

```

Main Essence here, if you working with a class that has the behaviour of list, collection, its better to use indexers.
