## Extension Methods

____

Again procedure

1. Our way of doing things, which is creating (custom) an extension method
2. The practical way we will almost always be USING extension methods, rather than creating one.


What is extension methods? To add methods to an existing class

1. Without changing its source code
2. Without creating a new class that inherits from it.

Lets take exmaples where we have string variable called post, with long string that we want to shorten


```
string post = "This is the start of a long blog post and now is the time for ....";
```

So in order to shorten the string we want to

- change the string class
- inherit from string class

BOTH of the above is not allowed

So what we want to do is to add a method to the string class, and from the above we see that it is not possible.

In order to CREATE an extension methods we use the following syntax



1. Start with name of class you want to extend
2. class must be static
3. methed must be static
4. first arg of method must be ```this type```,of the type you want to extend, AND give name. This will be the source string that we workng with.


And all the logic to get shortened string, that is number of words


Within logic, we the following:

```
var words = str.Split(' ');

```

When this happens we split all the words, separated by blank space into words array.

Adn when we look at the methods that is available on words array, we find ```words.Take(..)```, where Take is an extension method that was created by .NET. But array of which words is the type of, does not have a Take method??. But string has Take method. This is an example of USING an extension method, when we did the assignment code above (later we will see in more detail, why we USE extension methodsmm rather than creating it)

TAKE (extension) method was created by MS, that can be applied to aby class that implements the Enumerable interface, of which arrays does. This was done after the fact.
 
 
Because I am working in Jupyter Noteboo, I cannot creat a class for my extension (although this s the rigth approach:

```
(3,26): error CS1109: Extension methods must be defined in a top level static class; StringExtensions is a nested class
```

Outside this environment, we need to define it in a class.

In practice, its recommended to put the extension method into System namespace. Otherwise we will have all ```using...``` at top


This a good way to shortned strings


What if MS changes the classes(stirings) implementation, and create the same extension method as mine, which make be an instance method

And instance methods takes priority over statci methods. What MS suggests use extension methods only when you have to otherwise if you have source code modify it, or create a class that inherits from that class. 

````
//public static class StringExtensions{
    
    public static string Shorten(this String str, int numberOfWords){
        
        if (numberOfWords < 0)
            throw new ArgumentOutOfRangeException("Words should be greater than 0");
        if (numberOfWords == 0)
               return "";
        var words = str.Split(' ');
        
        
        if (words.Length <= numberOfWords)
            return str;
        
        words.Take(numberOfWords);
            
        return string.Join(" ", words.Take(numberOfWords));
        
    }
//}
````

````
string post = "This is the start of a long blog post and now is the time for ....";

var shortenedPost = post.Shorten(5);

Console.WriteLine(shortenedPost);
````

## Practcal terms you will be USING extension methods

_____

Lets explore...instead of creating it...

An example is the IEnumerable interface.

We create an instance of IEnumerable of type int called numbers and assign it to List of int

```
IEnumerable<int> numbers = new List<int>

```

List is generic, and IEnumerable is also generic

But what does the above assignment do? Now IEnumerable has avaiable lots of extension methods created by MS, that we can use which is defined in e.g. System.Linq namespace. And these exytension methods allow us to query objects.


Summary, very useful, but we only WILL be using it, rather than creating it.


````
string post = "This is the start of a long blog post and now is the time for ....";

var shortenedPost = post.Shorten(5);

IEnumerable<int> numbers = new List<int>(){5,6,719,15726,76,65};

var max = numbers.Max();

Console.WriteLine(shortenedPost);

Console.WriteLine("The maximum number is {0}",max);
````
